```python
from sklearn.preprocessing import normalize
```

```python
normalized=pd.DataFrame(normalize(data),columns=data.columns)
```

```python
normalized
```

```python
features=normalized[['Sorting_Time']]
```

```python
target=normalized[['Delivery_Time']]
```

```python
print(features)
print(target)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(features,target,train_size=0.75,random_state=100)
```

```python
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
liner.fit(xtrain,ytrain)
```

```python
ypredt=liner.predict(xtest)
```

```python
ypredt
```

```python
liner.score(xtest,ytest)
```

```python
error=ytest-ypredt
error
```
