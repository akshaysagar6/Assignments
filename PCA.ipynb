## KMEANS CLUSTERING

```python
from sklearn.cluster import KMeans
```

## optimum  value for kmeans

```python
output=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(new_df)
    output.append(kmeans.inertia_)
```

```python
output
```

```python
plt.plot(range(1,10),output)
plt.xlabel('Clusters')
plt.ylabel('Inertia')
plt.title('optimum value')
```
