```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestClassifier
random=RandomForestClassifier(n_estimators=21)
```

```python
df= pd.read_csv(r'C:\Users\aksha\Downloads\Company_Data (2).csv')
```

```python
df
```

```python
df.dtypes
```

```python
df['High'] = df['Sales'].map(lambda x: 1 if x>8 else 0)
```

```python
df.dtypes
```

```python

df['ShelveLoc']=df['ShelveLoc'].map({'Good':1,'Medium':2,'Bad':3})
```

```python
df
```

```python
df.dtypes
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
```

```python
df['Urban']=label.fit_transform(df['Urban'])
df['US']=label.fit_transform(df['US'])
```

```python
features=df.drop(['Sales','High'],axis=1)
```

```python
target=df[['High']]
```

```python
features
```

```python
target
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(features,target,train_size=0.75,random_state=100)
```

```python
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
print(ytest.dtypes)
```

```python
random.fit(xtrain,ytrain)
```

```python
ypredt=random.predict(xtest)
```

```python
ypredt
```

```python
accuracy_score(ytest,ypredt)
```

## Grid Search

```python
params={'n_estimators': np.arange(1,20)}
```

```python
grid=GridSearchCV(random,params)
```

```python
grid.fit(xtrain,ytrain)
```
