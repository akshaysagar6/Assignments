```python
target=df[['size_category']]
target
```

```python
target.value_counts()
```

```python
xtrain=df.drop('size_category',axis=1)
```

```python
xtrain
```

```python
string_var=["month","day"]
```

```python
label=LabelEncoder()
```

```python
for i in string_var:
    xtrain[i]=label.fit_transform(xtrain[i])
```

```python
xtrain
```

```python
xtrain.dtypes
```

```python
normal=Normalizer()
```

```python
df1=pd.DataFrame(normal.fit_transform(xtrain),columns=xtrain.columns)
```

```python
df1
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='rbf',gamma=0.2,C=1.0)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(df1,target,random_state=100,train_size=0.75)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.shape
```

```python
ytest.shape
```

```python
svc.fit(xtrain,ytrain)
```

```python
ypredt=svc.predict(xtest)
```

```python
ypredt
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,ypredt)
```

```python
params={'kernel':['rbf','linear','poly','sigmoid'],'gamma':[0.1,0.4,0.6],'C':np.arange(1,5)}
```

```python
grid=GridSearchCV(svc,params)
```

```python
grid.fit(xtrain,ytrain)
```

```python
grid.best_params_
```
